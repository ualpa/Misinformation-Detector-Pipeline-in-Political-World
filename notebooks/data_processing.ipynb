### **1. Imports**

```python
import pandas as pd
import re
import string
from datetime import datetime

# Citești datele brute salvate anterior
df = pd.read_csv("../data/raw/articles_raw.csv")
print(f"Articole încărcate: {len(df)}")
df.head()
```

### **2. Checking number of columns and null values**

```python
print(df.columns.tolist())
print("\nValori lipsă per coloană:")
print(df.isnull().sum())
```

### **3. Cleaning function**

```python
def clean_text(text):
    if not isinstance(text, str): 
        return ""
    
    text = text.lower()                                      
    text = re.sub(r'http\S+|www\S+', '', text)              
    text = re.sub(r'<.*?>', '', text)                        
    text = re.sub(r'@\w+', '', text)                         
    text = re.sub(r'#\w+', '', text)                         
    text = re.sub(r'\[.*?\]', '', text)                      
    text = re.sub(r'[^\w\s]', '', text)                      
    text = re.sub(r'\s+', ' ', text).strip()                 
    
    return text
```

### **4. Cleaning**

```python
df['title_clean'] = df['title'].apply(clean_text)
df['description_clean'] = df['description'].apply(clean_text)

df['full_text'] = df['title_clean'] + ' ' + df['description_clean']

df[['title', 'title_clean']].head(3)

df = df[df['full_text'].str.len() > 20].reset_index(drop=True)

df = df.drop_duplicates(subset='full_text').reset_index(drop=True)

df['publishedAt'] = pd.to_datetime(df['publishedAt']).dt.date

print(f"Articole după curățare: {len(df)}")
```

### **5. Saving the clean data**

```python
df.to_csv("../data/processed/articles_clean.csv", index=False)
print("Salvat în data/processed/articles_clean.csv")

# Preview final
df[['title_clean', 'full_text', 'publishedAt']].head()
```
