1. Setup

```python
import requests
import pandas as pd
import json
from dotenv import load_dotenv, find_dotenv
import os

load_dotenv(find_dotenv(), override=True) 
API_KEY = os.getenv("NEWS_API_KEY")
print(f"Cheia mea este: '{API_KEY}'")
```

2. Fetch political articles

```python
url = "https://newsapi.org/v2/everything"

params = {
    "q": "politics misinformation fake news",
    "language": "en",
    "sortBy": "publishedAt",
    "pageSize": 100,
    "apiKey": API_KEY
}

response = requests.get(url, params=params)
data = response.json()

print(f"Status: {data['status']}")
print(f"Status: {data}")
print(f"Total articole găsite: {data['totalResults']}")
```

3. Data visualization

```python
articles = data["articles"]
df = pd.DataFrame(articles)

df[["source", "title", "description", "publishedAt", "url"]].head()
```

4. Saving the raw data

```python
df.to_csv("../data/raw/articles_raw.csv", index=False)
print(f"Salvate {len(df)} articole în data/raw/articles_raw.csv")
```
